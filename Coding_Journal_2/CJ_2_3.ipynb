```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,1000)

width = 0.01
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins = bins, alpha = 0.5)
plt.ylabel("x per bin")
plt.xlabel("x")
plt.savefig("CJ_2_3_Fig.pdf", bbox_inches = "tight", facecolor = "white")
```
