```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.dirichlet(alpha = (2.0, 4.0, 8.0), size = 1000)

width = 0.01
histmin = np.floor(x)
histmax = np.ceil(x) + width

bins = np.arange(histmin[0,0],histmax[0,0],width)
plt.hist(x,bins = bins, alpha = 0.5)
plt.ylabel("x per bin")
plt.xlabel("x")
plt.savefig("CJ_2_4_Fig.pdf", bbox_inches = "tight", facecolor = "white")
```
