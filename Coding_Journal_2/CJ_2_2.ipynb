```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sine(x):
    return np.sin(x)
def cosine(x):
    return np.cos(x)
```

```python
x = np.linspace(0,1,100)
y1 = sine(x)
y2 = cosine(x)
```
